```python
import random

# Generate 1000 unique random numbers between 1 and 1000
range1 = random.sample(range(1, 1001), 1000)

# Generate 40000 unique random numbers between 1 and 40000
range2 = random.sample(range(1, 40001), 40000)

# Generate 80000 unique random numbers between 1 and 80000
range3 = random.sample(range(1, 80001), 80000)

# Generate 200000 unique random numbers between 1 and 200000
range4 = random.sample(range(1, 200001), 200000)

# Generate 1000000 unique random numbers between 1 and 1000000
range5 = random.sample(range(1, 1000001), 1000000)
```

```python
class Node:
    
    def __init__(self, v):
        self.val = v
        self.left = None
        self.right = None
        
def inorder_traversal(node):
        if node:
        
            inorder_traversal(node.left)
            print(node.val)
            inorder_traversal(node.right)

        
def insert( root, node):
    
    if root is None:
        return node
    
    if root.val > node.val:
        root.left = insert(root.left,node)

    elif root.val < node.val:
        root.right = insert(root.right,node)

    return root
```

```python
# for tree1
root1 = Node(range1[0])

for i in range1:
    insert(root1,Node(i))

root2 = Node(range2[0])

for i in range2:
    insert(root2,Node(i))

root3 = Node(range3[0])

for i in range3:
    insert(root3,Node(i))
    
root4 = Node(range4[0])

for i in range4:
    insert(root4,Node(i))
    
root5 = Node(range5[0])

for i in range5:
    insert(root5,Node(i))
    
```

```python
def BFS(goal,root):
    explored = []
    q = []
    q.append(root)


    while len(q)>0:

        #print(q)
        
        qq = [i.val for i in q]
#         print(qq)
        
        node = q.pop(0)
        
        explored.append(node.val)
        
#         print('POP: ',node.val)
#         print()

        if node.val == goal:
#             print('Result:  ',explored)
            break


        if node.left is not None:
            q.append(node.left)

        if node.right is not None:
            q.append(node.right)
```

```python
def DFS(goal,root):
    explored = []
    s = []
    s.append(root)


    while len(s)>0:

        
        
        ss = [i.val for i in s]
#         print(ss)
        
        node = s.pop()
        
        explored.append(node.val)
        
#         print('POP: ',node.val)
#         print()

        if node.val == goal:
#             print('Result:  ',explored)
            break
            
        if node.left is not None:
            s.append(node.left)

        if node.right is not None:
            s.append(node.right)
```
