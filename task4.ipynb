```python
def build_dcit(lis):
    di  = {}

    for i in lis:
        name = i[0]
        cost = i[1]
        di[name] = cost
    return di
```

```python
def my(x):
    return x[1]
```

```python
def UCS(d,goal):
    
    q = []
    q.append(('A',0))
    explored = []
    
    while len(q)>0:

        print('Queue :  ',q)
        node = q.pop(0)
        print('child pop: ',node) # node[0] --> node name AND node[1] --> cost
        
        if node[0] not in explored:
            explored.append(node[0])
        
        if node[0] == goal:
            return explored 
        
        print('Explored: ',explored)

        child = d.get(node[0])
        print('current child: ',child)
        
        for i in child:
            print(i)
            
            cost =  i[1] + node[1]
            name = i[0]
            new = (name,cost)

            
            print('{} , parent cost ({}) + child cost ({}) = {}'.format(i,node[1],i[1],cost))
            #print('////\\\\ ',new)
            
            
            if new[0] not in explored and new[0] not in build_dcit(q):
                
                q.append(new) 
                
            
                
            elif new[0] in build_dcit(q):
                #rint('[[[[[[[[[[[[[[[duplicate]]]]]]]]]]]]]]]')

                #rint(new[0],build_dcit(q)[new[0]])

                old_data = build_dcit(q)
                old_cost = old_data[new[0]]
                old_name = new[0]
                old = (old_name,old_cost)

                #rint(old)
                #rint('NNNNNNNNNNNNNNNNNNNNNN',new , node)
                if new[1] < old[1]:
                    #rint('^^^^^^^*******%%%%%%%% in if')
                    inde = q.index(old)
                    q[inde] = new
                else:
                    pass

        
        q = sorted(q,key= my)
        #print(q)
        print(50*'*')
        
    
UCS(dd,'E')
```
